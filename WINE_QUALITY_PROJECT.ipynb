```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor

import pandas as pd

file_path = "/content/drive/MyDrive/dataset/winequality-white.csv"
wine_data = pd.read_csv(file_path)
print(wine_data.head())

file_path = "/content/drive/MyDrive/dataset/winequality-red.csv"
wine_data = pd.read_csv(file_path)
print(wine_data.head())

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
wine_data = pd.read_csv(url, sep=';')
plt.figure(figsize=(8, 5))
sns.countplot(x='quality', data=wine_data)
plt.title('Distribution of Wine Quality Scores')
plt.xlabel('Quality Score')
plt.ylabel('Count')
plt.show()
plt.figure(figsize=(8, 5))
sns.scatterplot(x='alcohol', y='quality', data=wine_data)
plt.title('Relationship between Alcohol Content and Wine Quality')
plt.xlabel('Alcohol Content')
plt.ylabel('Quality')
plt.show()
summary_stats = wine_data.describe()
X = wine_data.drop('quality', axis=1)
y = wine_data['quality']
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X, y)
feature_importances = pd.Series(rf.feature_importances_, index=X.columns)
sorted_features = feature_importances.sort_values(ascending=False)
top_n = 5
print(f"Top {top_n} Most Important Features:")
print(sorted_features.head(top_n))
```
